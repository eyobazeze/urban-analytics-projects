## 2. Lists, tuples, dictionaries, sets

These are Python's core ways of holding multiple values. Each has a different purpose:

| Structure | Ordered? | Changeable? | Duplicates allowed? | Typical use |
|---|---|---|---|---|
| `list` | Yes | Yes | Yes | A sequence you'll modify — e.g. sub-city names |
| `tuple` | Yes | No | Yes | A fixed record — e.g. a single (lat, lon) coordinate |
| `dict` | Yes (3.7+) | Yes | Keys must be unique | Labeled data — e.g. `{"Bole": 328900}` |
| `set` | No | Yes | No duplicates | Unique membership checks — e.g. list of visited sub-cities with no repeats |

````python
# List — ordered, changeable
subcities = ["Bole", "Yeka", "Kirkos", "Lideta", "Arada"]
subcities.append("Gulele")
print(subcities)

# Tuple — fixed coordinate, shouldn't change
addis_center = (9.0192, 38.7525)  # (latitude, longitude)
print(addis_center)

# Dictionary — labeled data
population_by_subcity = {
    "Bole": 328900,
    "Yeka": 401000,
    "Kirkos": 235100
}
print(population_by_subcity["Yeka"])

# Set — unique values only
visited = {"Bole", "Yeka", "Bole", "Arada"}  # duplicate "Bole" is dropped
print(visited)
````

## 3. Control flow — `if`/`elif`/`else`, `for`, `while`

This is how your code makes decisions and repeats actions. Indentation (4 spaces, consistently) defines the blocks — Python doesn't use `{}` or `end` keywords.

````python
population = 401000

if population > 350000:
    print("High population sub-city")
elif population > 200000:
    print("Medium population sub-city")
else:
    print("Lower population sub-city")

# for loop over a list
subcities = ["Bole", "Yeka", "Kirkos"]
for city in subcities:
    print("Sub-city:", city)

# while loop
count = 0
while count < 3:
    print("Count is", count)
    count += 1
````

## 5. String formatting (f-strings)

## 6. File I/O — plain Python (no Pandas yet)

Before Pandas handles files for you, it's worth knowing what's actually happening: opening a file, reading lines, writing lines, closing it. The `with open(...) as f:` pattern automatically closes the file for you even if an error occurs.

````python
# Writing a simple CSV by hand (no libraries)
data = [
    ("Bole", 328900),
    ("Yeka", 401000),
    ("Kirkos", 235100)
]

with open("subcities_manual.csv", "w") as f:
    f.write("sub_city,population\n")   # header row
    for name, pop in data:
        f.write(f"{name},{pop}\n")

print("File written.")
````

````python
# Reading it back
with open("subcities_manual.csv", "r") as f:
    lines = f.readlines()

for line in lines:
    print(line.strip())   # .strip() removes the trailing newline character
````

## 7. Error handling — `try`/`except`

Real data breaks things — a missing field, a value that isn't actually a number, a file that doesn't exist. `try`/`except` lets your script handle that gracefully instead of crashing. This matters a lot once you're pulling live OSM data, which sometimes fails or times out.

````python
# Without error handling, this crashes the whole script:
values = ["100", "250", "abc", "400"]

# With error handling:
total = 0
skipped = 0

for v in values:
    try:
        total += int(v)
    except ValueError:
        print(f"Skipping invalid value: {v}")
        skipped += 1

print(f"Total: {total}, Skipped: {skipped} value(s)")
````

## 8. Basic OOP — classes & objects

You don't need to become an OOP expert, but you do need to *read* it, since GeoPandas objects (a `GeoDataFrame` is a class), OSMnx graphs, and most real libraries are built with classes. A class is a blueprint; an object is a specific instance made from that blueprint.

````python
class SubCity:
    def __init__(self, name, population, area_km2):
        self.name = name
        self.population = population
        self.area_km2 = area_km2

    def density(self):
        return round(self.population / self.area_km2, 1)

    def __repr__(self):
        return f"SubCity({self.name}, pop={self.population})"

bole = SubCity("Bole", 328900, 122.8)
yeka = SubCity("Yeka", 401000, 85.9)

print(bole)
print(bole.density())
print(yeka.density())
````

```python
class SubCity:
    def __init__(self, name, population, area_km2):
        self.name = name
        self.population = population
        self.area_km2 = area_km2

    def density(self):
        return round(self.population / self.area_km2, 1)

    def is_dense(self, threshold=3000):
        return self.density() > threshold

    def __repr__(self):
        return f"SubCity({self.name}, pop={self.population})"

bole = SubCity("Bole", 328900, 122.8)
yeka = SubCity("Yeka", 401000, 85.9)

print(bole.name, "dense?", bole.is_dense())
print(yeka.name, "dense?", yeka.is_dense())
```

**Why this matters later:** when you write `gdf.geometry.buffer(500)` in GeoPandas, `buffer` is a method exactly like `density()` above — defined once in a class, called on every object built from it. Once this clicks, library code stops looking like magic.
